4. Limpeza de Linhas Desnecessárias

```python
linhas_remover = ['Unit Symbol',
    "Detection Limit",
    "Analysis Method"
]
df_litogeoquimica_Results = df_litogeoquimica_Results[~df_litogeoquimica_Results.iloc[:, 0].astype(str).isin(linhas_remover)]

df_litogeoquimica_Results = df_litogeoquimica_Results.reset_index(drop=True)
```

5. Organização da Ordem das Colunas e Classificação

```python
# Ordem desejada das colunas
nova_ordem_colunas = ['Analyte Symbol', "LOI", "Total","Sc", "Co", "Ni", "Cr", "V",
 "Ba", "Rb", "Sr", "Y", "Zr", "Nb",
 "La", "Ce", "Pr", "Nd", "Sm", "Eu",
 "Gd", "Tb", "Dy", "Ho", "Er", "Tm", "Yb", "Lu",
 'U', 'Pb', 'Th',
 "SiO2", "TiO2", "Al2O3", "Fe2O3(T)", "MnO", "MgO",
 "CaO", "Na2O", "K2O", "P2O5", "Hf", "Ta"]
# Reordenar colunas do DataFrame conforme a nova ordem
df_litogeoquimica_results_ordenado = df_litogeoquimica_Results[nova_ordem_colunas]

# Mapeamento de classes
mapeamento_classes = {
    "Analyte Symbol": "",
    "Sc": "Compatíveis",
    "Co": "Compatíveis",
    "Ni": "Compatíveis",
    "Cr": "Compatíveis",
    "V": "Compatíveis",
    "Ba": "Incompatíveis móveis",
    "Rb": "Incompatíveis móveis",
    "Sr": "Incompatíveis móveis",
    "Y": "Incompatíveis imóveis",
    "Zr": "Incompatíveis imóveis",
    "Nb": "Incompatíveis imóveis",
    "La": "Elementos terras-raras",
    "Ce": "Elementos terras-raras",
    "Pr": "Elementos terras-raras",
    "Nd": "Elementos terras-raras",
    "Sm": "Elementos terras-raras",
    "Eu": "Elementos terras-raras",
    "Gd": "Elementos terras-raras",
    "Tb": "Elementos terras-raras",
    "Dy": "Elementos terras-raras",
    "Ho": "Elementos terras-raras",
    "Er": "Elementos terras-raras",
    "Tm": "Elementos terras-raras",
    "Yb": "Elementos terras-raras",
    "Lu": "Elementos terras-raras",
    "U": "Sem classe definida",
    "Pb": "Sem classe definida",
    "Th": "Sem classe definida"
}

# Criar nova linha 'Classe' com base no mapeamento
df_litogeoquimica_results_ordenado.loc['Classe'] = [mapeamento_classes.get(col, 'Desconhecido') for col in nova_ordem_colunas]

display(df_litogeoquimica_results_ordenado)

```
