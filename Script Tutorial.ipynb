#Seleção de Colunas Específicas de um DataFrame

* Para selecionar colunas específicas em um DataFrame, use duas sequências de colchetes '[]'. A primeira sequência é usada para acessar o DataFrame, e a segunda para especificar uma lista de colunas que você deseja selecionar.

1. Seleção das Colunas

* Planilha amostras: (Código da Amostra, Área, Ponto)











```python
df_amostras = df_amostras[['Código da Amostra', 'Área', 'Ponto']]
```

* Planilha macro: (Código da Amostra, Classificação, Grau de Alteração, Natureza)

```python
df_macro = df_macro[['Código da Amostra', 'Classificação', 'Grau de Alteração', 'Natureza']]
# theresa colocou coluna 'Outras Estruturas...36' não encontrei
```

2. Renomeação

* Evitar acentos, espaços e caracteres especiais em nomes de colunas é uma boa prática de programação, pois aumenta a compatibilidade e legibilidade do código.

```python
df_amostras = df_amostras.rename(columns={'Código da Amostra':'id_amostra'})
df_amostras = df_amostras.rename(columns={'Ponto':'id_pt'})
df_amostras = df_amostras.rename(columns={'Área':'area'})
```

```python
df_macro = df_macro.rename(columns={'Código da Amostra':'id_amostra'})
df_macro = df_macro.rename(columns={'Classificação':'lito'})
df_macro = df_macro.rename(columns={'Grau de Alteração':'grau_alt'})
df_macro = df_macro.rename(columns={'Natureza':'natureza'})
```

##Integrando os DataFrames

* Quando dois DataFrames compartilham uma coluna em comum, podemos uni-los para integrar informações.

1. Verificação de Duplicadas em amostras e macro

```python
duplicatas_amostras = df_amostras[df_amostras.duplicated(subset=['id_amostra'])]
print(duplicatas_amostras)
```

```python
duplicatas_macro = df_macro[df_macro.duplicated(subset=['id_amostra'])]
print(duplicatas_macro)
```

2. Verificar se todas as amostras de macro estão em amostras

```python
faltantes = df_macro[~df_macro['id_amostra'].isin(df_amostras['id_amostra'])]
```

3. Juntando os DataFrames

* Realização do precesso chamado left join para fundir 2 tabelas distintas, mas ainda mantendo linhas que possuem valores vazios pois ainda não possui correspondente na outra tabela. Isto é usado para manter a visualização dos dados faltantes.

```python
amostras_macro = pd.merge(df_amostras, df_macro, on = 'id_amostra', how = 'left')
```

4. Verificar o Resultado da Junção

* Após a junção, o DataFrame amostras_macro conterá todas as colunas de amostras e as colunas adicionais de macro onde houver correspondência pela coluna id_amostra

```python
amostras_macro.head(5)
```

##Exportando o novo DataFrame em XLSX

 * A função .to_excel() funciona para transformar o DataFrame amostras_macro em um arquivo Excel e salvando no caminho escolhido.

```python
amostras_macro.to_excel('/content/Script-Tutorial-Python/Dados/amostras_macro.xlsx')
```

##Parte 2 - Transformando o XLSX em formato Shapefile

##Instalação e Importação de Pacotes

1. Importação das bibliotecas

* Importando bliblioteca geopandas necessária para trabalhar com dados espaciais

```python
import geopandas as gpd
```

2. Importação dos arquivos xlsx e shp como DataFrames

* Importação dos arquivos shapefile onde possui as informações espaciais das amostras

```python
shp_pontos = gpd.read_file('/content/pontos.shp')
shp_amostras = gpd.read_file('/content/amostras.shp')
```

```python
shp_pontos.head(5)
```

##Seleção de Colunas Específicas de um DataFrame

1. Renomeação

* Modificação do nome da coluna 'Código do P' para 'id_pt' para possibilitar relacionar as colunas nos dois DataFrames

```python
shp_pontos = shp_pontos.rename(columns={'Código do P':'id_pt'})
```

2. Seleção das Colunas

* Para selecionar colunas específicas em um DataFrame, use duas sequências de colchetes '[]'.

```python
shp_pontos = shp_pontos[['id_pt','Latitude','Longitude','Northing (N','Easting (E)','Fuso','geometry']]
```

```python
shp_pontos
```

##Integrando os DataFrames

1. Juntando os DataFrames

* Realização do precesso chamado left join para fundir 2 tabelas distintas, mas ainda mantendo linhas que possuem valores vazios pois ainda não possui correspondente na outra tabela. Isto é usado para manter a visualização dos dados faltantes.


```python
amostras_macro_shp = pd.merge(amostras_macro, shp_pontos, on = 'id_pt', how = 'left')
```

```python
amostras_espacializadas = amostras_macro_shp[['id_amostra','id_pt','area','lito','grau_alt','natureza','Latitude','Longitude','Fuso','geometry']]
```

2. Definindo como GeoDataFrame

* Especificar a coluna que contém os dados espaciais no formato onde apresenta a latitude e longitude juntos

```python
amostras_espacializadas = gpd.GeoDataFrame(amostras_espacializadas, geometry="geometry")
```

* Especificação do sistema de referência de coordenadas

```python
amostras_espacializadas.set_crs(epsg=4326)
```

##Exportando o novo GeoDataFrame em shapefile

```python
amostras_espacializadas.to_file('/content/Script-Tutorial-Python/Dados/amostras_espacializadas.shp')
```
